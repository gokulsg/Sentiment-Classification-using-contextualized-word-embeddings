1) 

LSTM can solve the problems of standard RNN like learning long-term temporal dependencies and vanishing gradient problem. LSTM networks are a type of RNN that uses gates to maintain the flow of information. LSTMs deal with vanishing and exploding gradient problem by using these gates, which allow for a better control over the gradient flow and enable better preservation of long-range dependencies.

2) 

Static word embeddings:
A static word embedding is a function that maps each word type to a single vector. These vectors are typically dense represenatations and have much lower dimensionality than the size of the vocabulary. This mapping function typically ignores that the same string of letters may have different senses (dining table vs a table of contents) or parts of
speech (to table a motion vs a table). It produces the same embeddings for the words irrespective of the context.

example: Skip-Gram & CBOW (aka Word2Vec), Glove, fastText

Contextualized word embeddings:
Contextualized word embedding learn sequence-level semantics by considering the sequence of all words in the documents. They produce different representations for polysemous words based on their context. 

example: BERT, GPT, Transformer XL, ELMO

Reference: https://courses.engr.illinois.edu/cs546/sp2020/Slides/Lecture04.pdf

3)

Static Word Embeddings fail to capture polysemy. They generate the same embedding for the same word in different contexts. But, Contextualized words embeddings aim at capturing word semantics in different contexts to address the issue of polysemous and the context-dependent nature of words.

Static Word Embeddings could only leverage off the vector outputs from unsupervised models for downstream tasks. They were mostly shallow models and were often discarded after training (e.g. word2vec, Glove). The output of Contextualized Word Embedding training is the trained model and vectors — not just vectors.

Reference: https://towardsdatascience.com/from-pre-trained-word-embeddings-to-pre-trained-language-models-focus-on-bert-343815627598

4)

Transfer learning is a method in machine learning that focuses on storing knowledge gained while solving one problem and applying it to a different but related problem. For example, knowledge gained while learning to recognize cars could apply when trying to recognize trucks.

The BERT model is pretrained on masked words prediction task (Masked Language Modeling - MLM) and Next sentence prediction. Yes, we can pretrain the model without any task using unsupervised pre-training.

5)

The Attention mechanism enables the transformers based models to have extremely long term memory. A transformer model can “attend” or “focus” on all previous tokens that have been generated using this attention mechanism. Self-attention helps the model to learn the “understanding” of other relevant words into the one we’re currently processing.
